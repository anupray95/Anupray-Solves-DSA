```python
class TreeNode:
    def __init__(self, val=None, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

### Level order traversal, array to tree and vice-versa

```python
def arr_to_tree(arr):
    root = TreeNode(arr[0])
    queue = [root]
    i = 1 #because already processed root
    while len(queue) != 0 and i < len(arr):
        node = queue.pop(0)
        if i < len(arr) and arr[i] != None:
            child = TreeNode(arr[i])
            node.left = child
            queue.append(child)
        i += 1

        if i < len(arr) and arr[i] != None:
            child = TreeNode(arr[i])
            node.right = child
            queue.append(child)
        i += 1
    return root
```

```python
def tree_to_arr(root):
    queue = [root]
    result = []

    while len(queue):
        node = queue.pop(0)
        if(node):
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
            continue
    while result and result[-1] == None:
        result.pop()
        
    return result
```

```python
root = arr_to_tree([3,9,20,None,None,15,7])
```

```python
tree_to_arr(root)
```
