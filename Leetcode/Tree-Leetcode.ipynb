Problem: https://leetcode.com/problems/sum-root-to-leaf-numbers/

Solution: https://leetcode.com/problems/sum-root-to-leaf-numbers/submissions/1867790349/

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def func(self, root, result=0):
        # if no root return 0;
        # we should not return result here
        # because it will get added twice for leaf nodes
        # as left of leaf is None + right of leaf is None
        if not root:
            return 0
        # find the current value for this node
        result = result * 10 + root.val
        # if its a leaf node, return the result
        if not root.left and not root.right:
            return result

        # find sum of left subtree and right subtree
        sumLeft = self.func(root.left, result)
        sumRight = self.func(root.right, result)
        return sumLeft + sumRight

    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        # self.summ = 0
        # root.right = None
        return self.func(root)
        
```

## Range sum of Binary Search Tree

<span style="
    display:inline-flex;

Problem: https://leetcode.com/problems/range-sum-of-bst

Solution: https://leetcode.com/problems/range-sum-of-bst/submissions/1870467575

```python
class Solution:
    def func(self, root, low, high):
        if not root:
            return 0
        # if root value in range, then add it to result
        if root.val >= low and root.val <=high:
            self.result += root.val

        # if root value is greater than low, then go to left branch
        if root.val >= low:
            self.func(root.left, low, high)
        # if root value is smaller than high, then only go to right branch
        if root.val <= high:
            self.func(root.right, low, high)

        
    def rangeSumBST(self, root: Optional[TreeNode], low: int, high: int) -> int:
        self.result = 0
        self.func(root, low, high)
        return self.result
```

Question: https://leetcode.com/problems/binary-tree-postorder-traversal

Solution: https://leetcode.com/problems/binary-tree-postorder-traversal/submissions/1875606846

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        self.res = []
        def func(root):
            if not root:
                return
            func(root.left)
            func(root.right)
            self.res.append(root.val)
        func(root)
        return self.res
```

```python

class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        self.res = True
        def func(root):
            if not root: 
                return 0
            if root.left == None and root.right == None:
                return 1
            left = func(root.left)
            right = func(root.right)
            print(root.val, left, right)
            if abs(left - right) > 1:
                self.res = False
            return 1 + max(left, right)
        func(root)
        return self.res

        
```
