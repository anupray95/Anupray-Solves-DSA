Problem: https://leetcode.com/problems/sum-root-to-leaf-numbers/

Solution: https://leetcode.com/problems/sum-root-to-leaf-numbers/submissions/1867790349/

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def func(self, root, result=0):
        # if no root return 0;
        # we should not return result here
        # because it will get added twice for leaf nodes
        # as left of leaf is None + right of leaf is None
        if not root:
            return 0
        # find the current value for this node
        result = result * 10 + root.val
        # if its a leaf node, return the result
        if not root.left and not root.right:
            return result

        # find sum of left subtree and right subtree
        sumLeft = self.func(root.left, result)
        sumRight = self.func(root.right, result)
        return sumLeft + sumRight

    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        # self.summ = 0
        # root.right = None
        return self.func(root)
        
```

## Range sum of Binary Search Tree

Problem: https://leetcode.com/problems/range-sum-of-bst

Solution: https://leetcode.com/problems/range-sum-of-bst/submissions/1870467575

```python
class Solution:
    def func(self, root, low, high):
        if not root:
            return 0
        # if root value in range, then add it to result
        if root.val >= low and root.val <=high:
            self.result += root.val

        # if root value is greater than low, then go to left branch
        if root.val >= low:
            self.func(root.left, low, high)
        # if root value is smaller than high, then only go to right branch
        if root.val <= high:
            self.func(root.right, low, high)

        
    def rangeSumBST(self, root: Optional[TreeNode], low: int, high: int) -> int:
        self.result = 0
        self.func(root, low, high)
        return self.result
```

Question: https://leetcode.com/problems/binary-tree-postorder-traversal

Solution: https://leetcode.com/problems/binary-tree-postorder-traversal/submissions/1875606846

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        self.res = []
        def func(root):
            if not root:
                return
            func(root.left)
            func(root.right)
            self.res.append(root.val)
        func(root)
        return self.res
```

Problem: https://leetcode.com/problems/search-in-a-binary-search-tree/

Solution: https://leetcode.com/problems/search-in-a-binary-search-tree/submissions/1865138718/


```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def searchBST(self, root, val):
        """
        :type root: Optional[TreeNode]
        :type val: int
        :rtype: Optional[TreeNode]
        """
        if not root:
            return None
        if root.val == val:
            return root
        if root.val < val:
            return self.searchBST(root.right, val)
        else:
            return self.searchBST(root.left, val)
 
        
```

Problem: https://leetcode.com/problems/validate-binary-search-tree

Solution: https://leetcode.com/problems/validate-binary-search-tree/submissions/1880682835

```python
class Solution:
    # comapre two values and handle None
    def checkLessThan(self, value1, value2):
        if value1 == None or  value2 == None:
            return True
        if value1 < value2:
            return True
        return False
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        def func(root, low=None, high=None):
            if not root:
                return True
            value = root.val
            # current root value must be low < value < high, else return false
            if not self.checkLessThan(low, value) or not self.checkLessThan(value, high):
                return False
            # check for left and right node
            # As one false means returning value, so adding both of the function call by "and" operation
            return func(root.left, low, value) and func(root.right, value, high) 
        return func(root)
```

## 701. Insert into a Binary Search Tree

Problem: https://leetcode.com/problems/insert-into-a-binary-search-tree/

Solution: https://leetcode.com/problems/insert-into-a-binary-search-tree/submissions/1881678754

```python
class Solution:
    def insertIntoBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:
        if not root:
            return TreeNode(val)
        lastNode = root
        # find the node, where we have to insert
        def findNode(root, val):
            nonlocal lastNode
            if not root:
                return
            # Keep track of the last valid node
            lastNode = root
            rootVal = root.val
            # if root value is smaller than value, then move to right, else left
            if rootVal < val:
                findNode(root.right, val)
            else:
                findNode(root.left, val)
        # findNode, result is stored in "lastNode" variable
        findNode(root, val)
        currNode =  TreeNode(val)
        # Add currNode as a child of lastNode
        # Keep the children node of 'lastNode' as it is
        if val > lastNode.val:
            lastNode.right = currNode
        else:
            lastNode.left = currNode
        return root
```
