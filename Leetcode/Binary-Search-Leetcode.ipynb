Solution: https://leetcode.com/problems/sqrtx/submissions/1882868054

```python
class Solution:
    def binarySearch(self, val, low, high):
        while low <= high:
            mid = (low + high)//2
            square = mid * mid
            # if we found floor value, then return
            if square <= val and (mid+1) ** 2 > val:
                return mid 
            # adjust high and low for binary search
            elif square > val:
                high = mid-1
            else:
                low = mid + 1

    def mySqrt(self, x: int) -> int:
        # Handled base case
        if x == 0 or x == 1:
            return x
        result = self.binarySearch(x, 0, x // 2)
        return result
        
```
