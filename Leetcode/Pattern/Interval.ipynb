```python
class Solution:
    def minMeetingRooms(self, start, end):
        # consider start and end as individual array
        start.sort()
        end.sort()
        n =len(start)
        
        start_ptr = 0
        end_ptr = 0
        maxm = -1
        overlap = 0
        
        # for every entry add counter(overlap) to 1
        # for every exit substract overlap to 1
        while start_ptr < n:
            if start[start_ptr] < end[end_ptr]:
                overlap += 1
                start_ptr += 1
            else:
                end_ptr += 1
                overlap -= 1
            maxm = max(maxm, overlap)
        return maxm
```

```python
class Solution:
    def canAttend(self, arr):
        start = []
        end = []
        for interval in arr:
            start.append(interval[0])
            end.append(interval[1])
        start.sort()
        end.sort()
        
        start_ptr = 0
        end_ptr = 0
        n = len(arr)
        overlap = 0
        while start_ptr < n:
            if start[start_ptr] < end[end_ptr]:
                overlap += 1
                start_ptr += 1
                if overlap > 1:
                    return False
            else:
                overlap -= 1
                end_ptr += 1
        return True
        
```
