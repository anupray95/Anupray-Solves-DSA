https://leetcode.com/problems/assign-cookies/submissions/2037713203/

```python
class Solution(object):
    def findContentChildren(self, g, s):
        """
        :type g: List[int]
        :type s: List[int]
        :rtype: int
        """
        noOfChild = len(g)
        g.sort()
        s.sort()
        gIndex = 0
        for val in s:
            # For Each cookie, check if a child can have that
            if g[gIndex] <= val:
                gIndex += 1
            if gIndex == noOfChild:
                return gIndex
        return gIndex

        
```

## 860. Lemonade Change

https://leetcode.com/problems/lemonade-change/submissions/2037720697/

```python
class Solution(object):
    def lemonadeChange(self, bills):
        """
        :type bills: List[int]
        :rtype: bool
        """
        cashBal = {
            5:0,
            10:0,
            20: 0
        }
        for val in bills:
            # in case of 5, just add 5, handling that at the end
            # in case of 10, return 5, if available
            if val == 10:
                if cashBal[5] < 1:
                    return False
                cashBal[5] -= 1
            # in case of 20, check if return coins are available
            elif val == 20:
                if cashBal[10] >= 1 and cashBal[5] >= 1:
                    cashBal[10] -=1 
                    cashBal[5] -=1
                elif  cashBal[5] >= 3:
                    cashBal[5] -=3
                else:
                    return False
            cashBal[val] += 1
        return True
```
