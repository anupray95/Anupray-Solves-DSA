https://leetcode.com/problems/assign-cookies/submissions/2037713203/

```python
class Solution(object):
    def findContentChildren(self, g, s):
        """
        :type g: List[int]
        :type s: List[int]
        :rtype: int
        """
        noOfChild = len(g)
        g.sort()
        s.sort()
        gIndex = 0
        for val in s:
            # For Each cookie, check if a child can have that
            if g[gIndex] <= val:
                gIndex += 1
            if gIndex == noOfChild:
                return gIndex
        return gIndex

        
```

## 860. Lemonade Change
<span style="

">
    Easy
</span>

https://leetcode.com/problems/lemonade-change/submissions/2037720697/

```python
class Solution(object):
    def lemonadeChange(self, bills):
        """
        :type bills: List[int]
        :rtype: bool
        """
        cashBal = {
            5:0,
            10:0,
            20: 0
        }
        for val in bills:
            # in case of 5, just add 5, handling that at the end
            # in case of 10, return 5, if available
            if val == 10:
                if cashBal[5] < 1:
                    return False
                cashBal[5] -= 1
            # in case of 20, check if return coins are available
            elif val == 20:
                if cashBal[10] >= 1 and cashBal[5] >= 1:
                    cashBal[10] -=1 
                    cashBal[5] -=1
                elif  cashBal[5] >= 3:
                    cashBal[5] -=3
                else:
                    return False
            cashBal[val] += 1
        return True
```

## 122. Best Time to Buy and Sell Stock II
<span style="
    display:inline-flex;
    align-items:center;

    color:#2cbfb3;

    font-family: system-ui, -apple-system, BlinkMacSystemFont;
">
    Medium
</span>

https://leetcode.com/problems/best-time-to-buy-and-sell-stock-ii/submissions/2039706062/

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        canBuy = True
        buy = 0
        result = 0
        n = len(prices)
        for i in range(n-1):
            val = prices[i]
            nextVal = prices[i+1]
            #buy if current value is less than to next Value
            # and currently we are not holding any stock
            if val < nextVal and canBuy:
                buy = val
                canBuy = False
            # sell if we are not holding any stock
            # and current value is greater than value, I bought
            # also nextValue must be smaller than current value, as we can get more profit by selling next time
            elif canBuy == False and val > buy and nextVal < val:
                result += (val-buy)
                canBuy = True
                buy = 0
        # Handle the last element
        if len(prices) >1 and prices[n-1] >= prices[n-2] and canBuy == False:
            result += (prices[n-1] - buy)
        return result
```

## 56. Merge Intervals
<span style="
    display:inline-flex;
    align-items:center;
    padding:6px 16px;
    border-radius:999px;
    background-color:#eefdfb;
    color:#2cbfb3;
    font-size:18px;
    font-weight:500;
    font-family: system-ui, -apple-system, BlinkMacSystemFont;
">
    Medium
</span>

https://leetcode.com/problems/merge-intervals/submissions/2039785792

```python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        # sort based on the start value
        intervals.sort(key=lambda x:x[0])
        n = len(intervals)
        ans = [intervals[0]]
        i = 1
        prev = intervals[0]
        while i < n:
            curr = intervals[i]
            prev = ans[-1]
            # if there is a overlap, change endTime
            # as we already sorted based on the start time, no need to worry about it
            if curr[0] <= prev[1]:
                currEnd = max(curr[1], prev[1])
                ans[-1][-1] = currEnd
            else:
                # if no overlap, append to result array
                ans.append(curr)
            i+= 1
        return ans
                    
```

## 763. Partition Labels
<span style="
    display:inline-flex;
    align-items:center;
    padding:6px 16px;
    border-radius:999px;
    background-color:#eefdfb;
    color:#2cbfb3;
    font-size:18px;
    font-weight:500;
    font-family: system-ui, -apple-system, BlinkMacSystemFont;
">
    Medium
</span>

https://leetcode.com/problems/partition-labels/submissions/2039810151

```python
class Solution(object):
    def partitionLabels(self, s):
        """
        :type s: str
        :rtype: List[int]
        """
        d = {}
        n = len(s)
        # add the start and end of every character in dictionary
        for i in range(n):
            char = s[i]
            if char not in d:
                d[char] = [i, i]
            else:
                d[char][-1] = i
        result = []
        i = 0
        while i < n:
            curr = d[s[i]]
            start, end = curr
            # Check all the character and their last occurence
            # replace end accordingly
            j = start
            while j <= end:
                if d[s[j]][-1] > end:
                    end = d[s[j]][-1]
                j+= 1
            result.append((end-start)+ 1)
            i = end+1 
        return result

```
