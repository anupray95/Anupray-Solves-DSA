Problem: https://leetcode.com/problems/power-of-three/

Solution: https://leetcode.com/problems/power-of-three/submissions/1872294228

```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        if n == 1:
            return True
        if n < 1 or n % 3 != 0:
            return False        
        return self.isPowerOfThree(n//3)
```

Problem: https://leetcode.com/problems/power-of-four/

Solution: https://leetcode.com/problems/power-of-four/submissions/1879071431

```python
class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if n < 1:
            return False
        if n == 1.0:
            return True
        return self.isPowerOfFour(n/4)
```
