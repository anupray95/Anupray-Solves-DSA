https://leetcode.com/problems/climbing-stairs/


Solution: https://leetcode.com/problems/climbing-stairs/submissions/1892443159

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        dp = [None]*(n+1)
        dp[0] = 1
        dp[1] = 1
        # for each step,we can go from prev or prev-1 steps
        # so add both previous result and return
        for i in range(2, n+1):
            dp[i] = dp[i-1] + dp[i-2]
        return dp[n]
        
```

Solution: https://leetcode.com/problems/house-robber/submissions/1892460926

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        def recursiveFunc(arr, n):
            if n <= 0:
                return 0
            # if we are considering the current element, then take n-2 next as n-1, we can't take
            # or take n-1 instaed
            return max(arr[n-1] + func(arr, n-2),  func(arr, n-1))
        def func(arr, n):
            dp = [None]*(n+1)
            dp[0] = 0
            dp[1] = arr[0]
            for i in range(2,n+1):
                dp[i] = max(arr[i-1] + dp[i-2], dp[i-1])
            return dp[n]        

        return func(nums, len(nums))
```
