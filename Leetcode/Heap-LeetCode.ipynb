Solution: https://leetcode.com/problems/top-k-frequent-elements/submissions/1866466358/

```python
import heapq
from collections import Counter
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        #count and store each frequency in dictionary
        counter = dict(Counter(nums)) 
        heap = []
        arr = []
        for ele, freq in counter.items():
            myTuple = tuple([freq, ele]) # sort based on frequency
            heapq.heappush(heap, myTuple)
            # we need k elements, if there are more than K, 
            # pop the minimum, thats in top of heap
            if len(heap) > k: 
                heapq.heappop(heap)
    
        return [ele[1] for ele in heap]
```

```python
import heapq
class Solution:
    def nearlySorted(self, arr, k): 
        result = []
        heap = []
        for ele in arr:
            heapq.heappush(heap, ele)
            if len(heap) > k:
                item = heapq.heappop(heap)
                result.append(item)
        while len(heap):
            result.append(heapq.heappop(heap))
        for i in range(len(arr)):
            arr[i] = result[i]
        
```

Problem: https://leetcode.com/problems/k-closest-points-to-origin

Solution: https://leetcode.com/problems/k-closest-points-to-origin/submissions/1866564901/

```python
import math, heapq
class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        heap = []
        for point in points:
            x = point[0]
            y = point[1]
            # find the distance
            # compare based on distance, in max heap
            dist = math.sqrt((x ** 2) + (y**2))
            pointTuple = tuple(point)
            heapq.heappush(heap, tuple([-dist, pointTuple]) )
            # remove extra elements from heap
            if len(heap) > k:
                heapq.heappop(heap)
        # return remaining points in heap
        return [list(point[1]) for point in heap]
            

        
```

```python
import heapq
class Solution:
   def minCost(self, arr):
        heap = []
        result = 0
        for ele in arr:
           heapq.heappush(heap, ele)
        while len(heap) > 1:
            # pop the smallest two and add them
            first = heapq.heappop(heap)
            sec = heapq.heappop(heap)
            summ = first + sec
            # keep the result updated by adding the current sum
            result += summ
            heapq.heappush(heap, summ)
        return result

arr = [4,3,2,6]
# output : 29
```
